```python
import pandas as pd
from selenium import webdriver
from bs4 import BeautifulSoup
import requests
```

```python
def criaUrl(ano):
    a = str(ano)
    url = "https://pt.wikipedia.org/wiki/Campeonato_Brasileiro_de_Futebol_de_"+a
    url = "https://pt.wikipedia.org/wiki/Campeonato_Brasileiro_de_Futebol_de_1968_(Torneio_Roberto_Gomes_Pedrosa)"
    return url
```

```python
def montarTabela(url):
    r = requests.get(url)
    content = r.content
    soup = BeautifulSoup(content, 'html.parser')
    table = soup.find_all(name='table')
    return table
```

```python
def acharTabela(table):
    table_str = str(table)
    for i in range(len(pd.read_html(table_str))):
        if(len(pd.read_html(table_str)[i])>=28):
            df = pd.read_html(table_str)[i]
            if(df.columns[0]=="Pos." or df.columns[0]=="Pos"):
                break
    return df
```

```python
def acharTabelaAlt(table,x):
    table_str = str(table)
    df = pd.read_html(table_str)[x]
    return df
```

```python
def preencherAno(df,ano):
    lista = []
    for i in range(len(df)):
        lista.append(ano)
    df['Ano'] = lista
```

```python
ano = 1968
url = criaUrl(ano)
table = montarTabela(url)
df = acharTabelaAlt(table,4)
```

```python
preencherAno(df,ano)
```

```python
df2 = acharTabelaAlt(table,6)
```

```python
preencherAno(df2,ano)
```

```python
result = pd.concat([df,df2])
```

```python
result = result.sort_values(by=['Pts','V','SG','GP'], ascending=False)
```
